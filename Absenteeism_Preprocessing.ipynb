```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
reason_columns
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df.head()
```

```python
df_concanated = df.copy()
```

```python
df_concanated.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df_concanated.columns = column_names
df_concanated
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_concanated = df_concanated[column_names_reordered]
df_concanated.head()
```

```python
df_reason_mod = df_concanated.copy()
```

```python
type(df_reason_mod['Date'])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'],format='%d/%m/%Y')
df_reason_mod['Date']
```

Day Of week

```python
def day_of_week(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day_of_week'] = df_reason_mod['Date'].apply(day_of_week)
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'],axis=1)
df_reason_mod.head()
```

```python
df_reason_mod.columns.values
```

```python
column_names=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Day_of_week']
```

```python
df_reason_mod.columns = column_names
df_reason_mod.head()
```

```python
df_reason_mod['Education'] = df_reason_mod['Education'].map({1:0,2:1,3:1,4:1})
```
